```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def F(n, x):
    return (4/n**2 * (-1)**n) * np.cos(n  * x)
```

```python
x = np.linspace(-2,0,50)
n_max = 5
F_ = np.pi**2/3
for i in range(1,n_max+1):
    F_ = F_ + F(i, x)
        
plt.plot(x, x**2, label='x**2')
plt.plot(x, F_, label='Фурье')
plt.legend()
```
