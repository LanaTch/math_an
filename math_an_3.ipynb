### Задание 2
Исследовать функцию $f(x)=x^3-x^2$ по плану

```python
x = np.linspace(-0.7, 1, 201)
plt.plot(x, x**3-x**2)
plt.grid(True)
plt.show()
```

1) Область задания и область значений

$dom(f)=\mathbb{R}$, $ran(f)=\mathbb{R}$

2) Нули функции и их кратность

$$f(x)=x^3-x^2=0$$
$$x^2(x-1)=0$$
значит, $x_1=0$ - нуль функции кратности 2 и $x_2=1$ - нуль функции кратности 1

3) Отрезки знакопостоянства

$N=\{x|x^3-x^2<0\}=(-\infty,0)\cup(0,1)$

$P=\{x|x^3-x^2>0\}=(1,\infty)$

4) Интервалы монотонности

чтобы найти локальные экстремумы, надо решить уравнение $f'(x)=0$

значит: $(x^3-x^2)'=3x^2\cdot 2x - 2x=2x\cdot(3x^2-1)=0$

локальный максимум $x=0$

локальный минимум $x=\sqrt{\frac{1}{3}}$

значит:

неубывание на M: $\forall x \in M:x<y\Rightarrow f(x)\le f(y)$ и $M=\left ( -\infty, 0\right ] \cup \left [\sqrt{\frac{1}{3}}, \infty \right)$

убывание на N: $\forall x \in N:x<y\Rightarrow f(x)>f(y)$ и $N=\left(0,\sqrt{\frac{1}{3}}\right)$

5) Четность функции.

функция не обладает свойством четности/нечетности

6) Ограниченность

функция не ограничена

7) Периодичность

функция непериодична
